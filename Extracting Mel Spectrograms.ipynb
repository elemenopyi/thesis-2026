```python
import os
import librosa
import IPython.display as ipd
import matplotlib.pyplot as plt
```

## Loading audio files with Librosa

```python
BASE_FOLDER = "audio/"
# scale_file = os.path.join(BASE_FOLDER, "1_amr-eng_4.wav")
scale_file = "audio/1_asm-eng_1.wav"
scale_file2 = "audio/1_asm-eng_7.wav"
scale_file3 = "audio/2_asm-eng_1.wav"
scale_file4 = "audio/2_asm-eng_7.wav"
```

```python
ipd.Audio(filename=scale_file)
```

```python
# load audio files with librosa
scale, sr = librosa.load(scale_file)
duration = librosa.get_duration(y=scale, sr=sr)
print(duration)
```

## Mel filter banks

```python
filter_banks = librosa.filters.mel(n_fft=1024, sr=16000, n_mels=128)
```

```python
filter_banks.shape
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(filter_banks, 
                         sr=sr, 
                         x_axis="linear")
plt.colorbar(format="%+2.f")
plt.show()
```

## Extracting Mel Spectrogram

```python
mel_spectrogram = librosa.feature.melspectrogram(y=scale, sr=sr, n_fft=1024, hop_length=44, n_mels=128)
```

```python
mel_spectrogram.shape
```

```python
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
```

```python
log_mel_spectrogram.shape
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(log_mel_spectrogram, 
                         x_axis="time",
                         y_axis="mel", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
import matplotlib.pyplot as plt
import librosa
import librosa.display

scale2, sr = librosa.load(scale_file2, sr=None)

mel_spectrogram = librosa.feature.melspectrogram(
    y=scale2,
    sr=sr,
    n_fft=1024,
    win_length=882,
    # win_length=320,
    hop_length=176,
    # hop_length=64,
    n_mels=128

)

log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)

print("duration:", librosa.get_duration(y=scale2, sr=sr))
print("shape:", log_mel_spectrogram.shape)

plt.figure(figsize=(10, 4))

librosa.display.specshow(
    log_mel_spectrogram,
    x_axis="time",
    y_axis="mel",
    sr=sr,
    hop_length=176
)

plt.colorbar(format="%+2.f dB")
plt.tight_layout()
plt.show()
```
